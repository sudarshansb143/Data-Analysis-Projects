```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings as ws
ws.filterwarnings("ignore")
```

```
temp = sns.load_dataset("tips")
```

```
temp.head()
```

```
plt.figure(figsize = (6,6))
sns.set()
# hue param should be added with the catagorial var
sns.scatterplot(x="total_bill", y = "tip", data= temp, hue = "sex", hue_order=["Male","Female"])
plt.show()
```

```
plt.figure(figsize = (6,6))
sns.set()
# Easy to compare the two catagorial vars w. r. to one another
# hue param should be added with the catagorial var
sns.countplot(temp["sex"], hue = temp["smoker"])
plt.show()
```

Conclusion from above plot is 
In males normally the 60 males are smoker and 98 males are non smoker.
In Females normally the  35 females are smoker and 58 females are non smoker

```
sns.set_style("white")
sns.relplot(x="total_bill", y ="tip", kind = "scatter", data = temp ,hue ="sex", palette="bright", col = "smoker", row = "day", legend="brief" )
plt.show()
```

From the above plot we can conclude that the scatter plot is firstly drawn with the continous vars and then for  convinience of for furthur analysis we add the catagorical vars to compare the  relationship in between them
